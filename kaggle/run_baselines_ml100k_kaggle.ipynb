# Run Baselines on MovieLens-100K (Kaggle)

This notebook runs the baseline pipeline through the shell script at kaggle/run_all_baselines_ml100k.sh.

```python
import os
import subprocess

REPO_URL = "https://github.com/faroukq1/LLM-Sequential-Recommendation.git"
REPO_DIR = "/kaggle/working/LLM-Sequential-Recommendation"

if not os.path.exists(REPO_DIR):
    subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)

os.chdir(REPO_DIR)
print("Working directory:", os.getcwd())
```

```python
import os
import subprocess

WORK_DIR = "/kaggle/working/llmseqrec_ml100k"

env = os.environ.copy()
env["WORK_DIR"] = WORK_DIR
env.setdefault("CORES", "2")
env.setdefault("EVAL_CORES", "1")
env.setdefault("NUM_EPOCHS", "6")
env.setdefault("TOP_KS", "10,20")
env.setdefault("MODELS", "Popular,SKNN,S-SKNN,SF-SKNN,V-SKNN,GRU4Rec,BERT4Rec,SASRec")

subprocess.run(["bash", "kaggle/run_all_baselines_ml100k.sh"], check=True, env=env)
print("Artifacts folder:", WORK_DIR)
```

```python
import os
import pandas as pd

results_path = "/kaggle/working/llmseqrec_ml100k/results/baseline_results_ml100k.csv"
failures_path = "/kaggle/working/llmseqrec_ml100k/results/baseline_failures_ml100k.csv"

if os.path.exists(results_path):
    results_df = pd.read_csv(results_path)
    display(results_df)
else:
    print("Results file not found:", results_path)

if os.path.exists(failures_path):
    failures_df = pd.read_csv(failures_path)
    if len(failures_df) > 0:
        display(failures_df)
```
