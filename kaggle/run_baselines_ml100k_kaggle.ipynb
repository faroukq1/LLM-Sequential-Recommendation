# Run Baselines on MovieLens-100K (Kaggle)

This notebook runs the baseline pipeline through the shell script at kaggle/run_all_baselines_ml100k.sh.

```python
import os
import subprocess

REPO_URL = "https://github.com/faroukq1/LLM-Sequential-Recommendation.git"
REPO_DIR = "/kaggle/working/LLM-Sequential-Recommendation"

if not os.path.exists(REPO_DIR):
    subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)

os.chdir(REPO_DIR)
subprocess.run(["git", "remote", "set-url", "origin", REPO_URL], check=True)
subprocess.run(["git", "fetch", "origin"], check=True)
current_branch = subprocess.check_output(
    ["git", "rev-parse", "--abbrev-ref", "HEAD"],
    text=True,
).strip()
subprocess.run(["git", "pull", "--ff-only", "origin", current_branch], check=True)

print("Working directory:", os.getcwd())
print("Updated branch:", current_branch)
```

```python
import os
import subprocess

WORK_DIR = "/kaggle/working/llmseqrec_ml100k"

# Preflight check: ensure required fixes are present in the repo checkout.
with open("kaggle/run_baselines_ml100k.py", "r", encoding="utf-8") as f:
    runner_src = f.read()
if "--sknn-k" not in runner_src or "--sknn-sample-size" not in runner_src:
    raise RuntimeError(
        "Your checkout has an old kaggle/run_baselines_ml100k.py. "
        "Run Cell 2 again to pull latest from your fork, or push the latest fix to your fork."
    )

with open("main/neural_model.py", "r", encoding="utf-8") as f:
    neural_src = f.read()
if "_train_on_numpy_batches" not in neural_src:
    raise RuntimeError(
        "Your checkout has an old main/neural_model.py without the neural Keras fallback. "
        "Run Cell 2 again to pull latest from your fork, or push the latest fix to your fork."
    )

env = os.environ.copy()
env["WORK_DIR"] = WORK_DIR
env.setdefault("CORES", "2")
env.setdefault("EVAL_CORES", "1")
env.setdefault("NUM_EPOCHS", "6")
env.setdefault("TOP_KS", "10,20")
env.setdefault("MODELS", "Popular,SKNN,S-SKNN,SF-SKNN,V-SKNN,GRU4Rec,BERT4Rec,SASRec")

subprocess.run(["bash", "kaggle/run_all_baselines_ml100k.sh"], check=True, env=env)
print("Artifacts folder:", WORK_DIR)
```
